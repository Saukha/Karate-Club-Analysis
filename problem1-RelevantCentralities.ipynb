```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as ny
from networkx.algorithms import community
from operator import itemgetter


print('Graph H: Original graph')
H=nx.read_graphml("problem1-ZacharysKarateClub.graphml")
print(nx.info(H))
density=nx.density(H)
print ("Network density = ", density)
print('The graph is connected? ', nx.is_connected(H))


# calculate centrality metrics for each node 
degree_centrality_dict = nx.degree_centrality(H) # degree centrality
#_degree_centrality_dict = nx.in_degree_centrality(H) # in-degree centrality, but graph is undirected
betweenness_centrality_dict = nx.betweenness_centrality(H)  # betweennes centrality
closeness_centrality_dict = nx.closeness_centrality(H) # closeness centrality
eigenvector_centrality_dict = nx.eigenvector_centrality(H)  # eigenvactor centrality
page_rank_dict = nx.pagerank(H) # page rank centrality

'''
# add centrality metrics to the node attribute
nx.set_node_attributes(H, degree_centrality_dict, 'degree')
#nx.set_node_attributes(H, in_degree_centrality_dict, 'in-degree') # but graph is undirected
nx.set_node_attributes(H, betweenness_centrality_dict, 'betweenness')
nx.set_node_attributes(H, closeness_centrality_dict, 'closeness')
nx.set_node_attributes(H, eigenvector_centrality_dict, 'eigenvector')
'''

#sort node_dict by metrality metric
sorted_degree = sorted(degree_centrality_dict.items(), key=itemgetter(1), reverse=True) 
print("Nodes sorted by degree centrality:")
for b in sorted_degree[:]:
    print(b)
print('  ')
sorted_betweenness = sorted(betweenness_centrality_dict.items(), key=itemgetter(1), reverse=True) 
print("Nodes sorted by betweenness centrality:")
for b in sorted_betweenness[:]:
    print(b)
print('  ')
sorted_closeness = sorted(closeness_centrality_dict.items(), key=itemgetter(1), reverse=True) 
print("Nodes sorted by closeness centrality:")
for b in sorted_closeness[:]:
    print(b)
print('  ')
sorted_eigenvector = sorted(eigenvector_centrality_dict.items(), key=itemgetter(1), reverse=True) 
print("Nodes sorted by eigenvector centrality:")
for b in sorted_eigenvector[:]:
    print(b)
print('  ')
sorted_page_rank = sorted(page_rank_dict.items(), key=itemgetter(1), reverse=True) 
print("Nodes sorted by page rank:")
for b in sorted_page_rank[:]:
    print(b)
print('  ')    

'''
communities_generator = community.girvan_newman(H,most_valuable_edge=1)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
sorted(map(sorted, next_level_communities))
print(sorted(map(sorted, next_level_communities)))
print('============')
'''
```
